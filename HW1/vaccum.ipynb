```python
class vaccum:
    def __init__(self, Envlist, agent_pos):
        self.Env = Environment(Envlist[0], Envlist[1])
        self.agent = agent(agent_pos)
        self.sum_score = 0
    
    def process(self):
        for i in range (0,1000):
            self.sum_score += self.Env.performance()
            st = self.agent.percept(self.Env.env)
            action = self.agent.action(st)
            self.Env.change(action, self.agent.locat)
            self.agent.move(action)
            #print(self.Env.env)
            
    def result(self):
        print(self.Env.ori_env, self.agent.ori_loc, self.sum_score)
        return

class Environment:
    def __init__(self, A_st, B_st):
        self.A_st = A_st
        self.B_st = B_st
        self.env = [A_st, B_st]
        self.ori_env = [A_st, B_st]
    
    def performance(self):
        score=0
        if self.A_st == 'clean':
            score += 1
        if self.B_st == 'clean':
            score += 1
        return score
    
    def change(self, action, pos):
        if action == 'suck':
            if pos == 'A':
                self.A_st = 'clean'
                self.env = [self.A_st, self.B_st]
            else:
                self.B_st = 'clean'
                self.env = [self.A_st, self.B_st]


class agent:
    def __init__(self,locat):
        self.locat = locat
        self.ori_loc = locat
        
    def pos(self):
        return self.locat
    
    def percept(self, env):
        if self.locat == 'A':
            return env[0]
        else:
            return env[1]
        
    def action(self, state):
        if state == 'dirty':
            return 'suck'
        elif state == 'clean':
            if self.locat == 'A':
                return 'right'
            elif self.locat == 'B':
                return 'left'
            
    def move(self, action):
        if action == 'left':
            self.locat = 'A'
        elif action == 'right':
            self.locat = 'B'
            
# Test
vac1 = vaccum(['clean','clean'], 'A'); vac1.process(); vac1.result()
vac2 = vaccum(['clean','dirty'], 'A'); vac2.process(); vac2.result()
vac3 = vaccum(['dirty','clean'], 'A'); vac3.process(); vac3.result()
vac4 = vaccum(['dirty','dirty'], 'A'); vac4.process(); vac4.result()
vac5 = vaccum(['clean','clean'], 'B'); vac5.process(); vac5.result()
vac6 = vaccum(['clean','dirty'], 'B'); vac6.process(); vac6.result()
vac7 = vaccum(['dirty','clean'], 'B'); vac7.process(); vac7.result()
vac8 = vaccum(['dirty','dirty'], 'B'); vac8.process(); vac8.result()
```
