```python
# 8-puzzle problem
# A_star & RBFS

class Board:
    def __init__(self, state, goal, count=0, parent=None):
        self.state = state   #[0-9]
        self.goal = goal    #[0-9]
        self.count = count
        self.score = self.f_func()
        self.parent = parent
    
    def checkParity(self):
        nstate = self.state.copy()
        nstate.remove(0)
        ngoal = self.goal.copy()
        ngoal.remove(0)
        count_st=0
        count_gl=0
        for i in range(1,len(nstate)+1):
            for j in range(i+1, len(nstate)+1):
                if nstate.index(i)>nstate.index(j):
                    count_st+=1
        for i in range(1,len(ngoal)+1):
            for j in range(i+1, len(ngoal)+1):
                if ngoal.index(i)>ngoal.index(j):
                    count_gl+=1
        #print(count_st, count_gl)
        if count_st%2 == count_gl%2:
            return True
        else:
            return False
    
    def change(self, count):
        exp_list=[]
        blank = self.state.index(0)
        if(blank not in [6,7,8]):
            nstate = self.state.copy()
            nstate[blank], nstate[blank + 3] = nstate[blank + 3], nstate[blank]
            nB=Board(nstate, self.goal, count, self)
            exp_list.append(nB)
        if(blank not in [2,5,8]):
            nstate = self.state.copy()
            nstate[blank], nstate[blank + 1] = nstate[blank + 1], nstate[blank]
            nB=Board(nstate, self.goal, count, self)
            exp_list.append(nB)
        if(blank not in [0,1,2]):
            nstate = self.state.copy()
            nstate[blank], nstate[blank - 3] = nstate[blank - 3], nstate[blank]
            nB=Board(nstate, self.goal, count, self)
            exp_list.append(nB)
        if(blank not in [0,3,6]):
            nstate = self.state.copy()
            nstate[blank], nstate[blank - 1] = nstate[blank - 1], nstate[blank]
            nB=Board(nstate, self.goal, count, self)
            exp_list.append(nB)
        return exp_list

    def ManhattanDist(self):
        dis = 0
        for i in range(0, 9):
            sti = self.state.index(i)
            goali = self.goal.index(i)
            x1 = sti%3
            y1 = sti//3
            x2 = goali%3
            y2 = goali//3
            xdis = x1-x2 if x1-x2>0 else x2-x1
            ydis = y1-y2 if y1-y2>0 else y2-y1
            dis+= (xdis + ydis)
        return dis
    
    def h_func(self):
        return self.ManhattanDist()
    
    def g_func(self):
        return self.count
    
    def f_func(self):
        return self.g_func()+ self.h_func()

    def isreach(self):
        if(self.state==self.goal):
            return True
        else:
            return False
    
class agent:
    def __init__(self, state, goal):
        self.board = Board(state, goal)
        
    def isvalid(self):
        return self.board.checkParity()
        
    def getPath(self, state):
        path = [state]
        while state.parent:
            state = state.parent
            path.append(state)
            #path=path[::-1]
        return path
    
    def A_star(self):
        stack=[]
        stack.append(self.board)
        visited = []
        count=0
        xxx=[]
        
        iii=0
        while(len(stack)!=0):
            min=999999
            for i in range(len(stack)):
                #print(stack[i].f_func())
                if stack[i].f_func()<min:
                    min=stack[i].f_func()
                    index= i
            st= stack[index]
            
            visited.append(st.state)
            
            if st.state in xxx:
                ww=True
            else:
                ww=False
            xxx.append(st.state)
            
            #print(iii, st.state, st.f_func(), st.g_func(), st.h_func(),ww)
            
            if st.isreach():
                path= self.getPath(st)
                path = path[::-1]
                return path
            
            count+=1
            for sti in st.change(count):
                if sti.state not in visited:
                    stack.append(sti)
            iii+=1
            #print(visited)
            del stack[index]
            
        return None
    
    def RBFS(self, node, flimit, count):
        # print(node.setup)
        if node.state == self.board.goal:
            path = self.getPath(node)
            path= path[::-1]
            for state in path:
                print(state.state)
            return node, flimit

        successors = []
        count += 1
        for state in node.change(count):
            successors.append(state)

        if len(successors) == 0:
            return None, 999

        for s in successors:
            s.score = max(s.f_func(), node.score)

        while True:
            successors.sort(key=lambda x: x.score)
            # for s in successors:
            #     print(s.score)
            # print("=" * 30)
            best_succ = successors[0]
            if best_succ.score > flimit:
                node.score = best_succ.score
                # print(best_succ.score)
                return None, best_succ.score
            if len(successors) > 1:
                alter = successors[1].score
            else:
                alter = 999
            result, bestf = self.RBFS(best_succ, min(flimit, alter), count)
            if result is not None:
                return result, bestf



start=[3,1,2,4,0,7,6,8,5]
start1=[3,1,2,4,0,5,6,7,8]
goal=[0,1,2,3,4,5,6,7,8]

ag=agent(start1,goal)
if ag.isvalid():
    
    print('A*:')
    w= ag.A_star()
    for i in range(0,len(w)):
        print(w[i].state)
    
    print('RBFS:')
    x, y=ag.RBFS(ag.board, 99, 0)
    
    
else:
    print("can't reach! ")
findsolution_Astar()
```
