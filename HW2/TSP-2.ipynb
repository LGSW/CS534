```python
# TSP problem 
# Local search: 
# hill-clinbing & generatic

import random
import math

class Cities:
    def __init__(self, n, ci, dist=None, visited= None, unvisited= None, cost=0, parent=None):
        self.amount= n
        self.ci = ci
        self.cost = cost
        self.parent = parent
        if visited!=None:
            self.visited=visited
        else:
            self.visited=[ci]
        
        if unvisited!=None:
            self.unvisited=unvisited
        else:
            self.unvisited=[i for i in range(0,n)]
            self.unvisited.remove(ci)
        self.cityid = [i for i in range(0,n)]
        #self.city.append([random.uniform(0,100), random.uniform(0,100)])
        
        if dist!=None:
            self.dist= dist 
        else:    
            self.dist=[([0]*n) for i in range(n)]
            for i in range(n):
                for j in range(n):
                    if i<j:
                        self.dist[i][j]= random.randint(0,10)
                    elif i>j:
                        self.dist[i][j]= self.dist[j][i]
                    else:
                        self.dist[i][j]= -1
                  
    def show_graph_dist(self):
        for i in range(self.amount):
            for j in range(self.amount):
                print(self.dist[i][j],end=' ')
                if(j==self.amount-1):
                    print()
    
    def isreach(self):
        if(self.unvisited==[]):
            return True
        else:
            return False
    
    
        
class agent:
    def __init__(self, n, ci):
        self.TSP = Cities(n, ci)
    
    def getPath(self, state):
        path = [state]
        while state.parent:
            state = state.parent
            path.append(state)
            #path=path[::-1]
        return path
    
    def calFinalCost(self, path):
        length= len(path)
        result= self.TSP.dist[path[0].ci][path[length-1].ci]
        a=path[0]
        for i in range(0,len(path)-1):
            a,b= path[i].ci, path[i+1].ci
            result+= self.TSP.dist[a][b]
        return result
    
    def calFinalCost_list(self, list):
        length= len(list)
        result= self.TSP.dist[list[0]][list[length-1]]
        for i in range(0, length-1):
            a, b = list[i], list[i+1]
            result+= self.TSP.dist[a][b]
        return result
    
    def hill_climbing(self, iterat):
        amount= self.TSP.amount
        st= [i for i in range(amount)]
        random.shuffle(st)
        cost= self.calFinalCost_list(st)
        v=0
        for i in range(iterat):
            #print(st)
            perm=[i for i in range(1,amount-1)]
            cut= random.sample(perm,2)
            cut1, cut2=min(cut[0],cut[1]), max(cut[0],cut[1])
            a, b, c= st[0:cut1],st[cut1:cut2],st[cut2:amount]
            new_ar=[]
            new_ar.append(a+b+c)
            new_ar.append(a+c+b)
            new_ar.append(b+a+c)
            new_ar.append(b+c+a)
            new_ar.append(c+a+b)
            new_ar.append(c+b+a)
            index=0
            for j in range(len(new_ar)):
                cos=self.calFinalCost_list(new_ar[j])
                if cos< cost:
                    cost=cos
                    index=j
            st=new_ar[index]
            if(index==0):
                v+=1
            else:
                v=0
            if v>=3:
                return st
        return st
            
    def genetic(self, iterat):
        amount= self.TSP.amount
        st=[]
        st1= [i for i in range(amount)]
        st2= [i for i in range(amount)]
        st3= [i for i in range(amount)]
        st4= [i for i in range(amount)]
        random.shuffle(st1)
        random.shuffle(st2)
        random.shuffle(st3)
        random.shuffle(st4)
        mincost=999999
        minpath=[]
        for i in range(iterat):
            cos1= self.calFinalCost_list(st1)
            cos2= self.calFinalCost_list(st2)
            cos3= self.calFinalCost_list(st3)
            cos4= self.calFinalCost_list(st4)
            
            mincost=min(cos1, cos2, cos3, cos4)
            if(cos1== mincost):
                minpath=st1
            elif(cos2== mincost):
                minpath=st2
            elif(cos3== mincost):
                minpath=st3
            else:
                minpath=st4
            
            
            p1, p2, p3, p4= 1/cos1, 1/cos2, 1/cos3, 1/cos4
            sum=p1+p2+p3+p4
            p1, p2, p3, p4= p1/sum, p2/sum, p3/sum, p4/sum
            pp1, pp2, pp3, pp4= p1, p1+p2, p1+p2+p3, 1
            ls=[]
            for j in range(0,4):
                rd= random.random()
                if(rd<pp1):
                    ls.append(st1)
                elif(rd>=pp1 and rd<pp2):
                    ls.append(st2)
                elif(rd>=pp2 and rd<pp3):
                    ls.append(st3)
                else:
                    ls.append(st4)
            cut= random.randint(1, amount-1)
            a1, a2= ls[0][0:cut], ls[0][cut:amount]
            b1, b2= ls[1][0:cut], ls[1][cut:amount]
            c1, c2= ls[2][0:cut], ls[2][cut:amount]
            d1, d2= ls[3][0:cut], ls[3][cut:amount]
            
            newl=[]
            newl.append(a1+b2)
            newl.append(b1+a2)
            newl.append(c1+d2)
            newl.append(d1+c2)
            for j in range(len(newl)):
                gene=newl[j][cut:]
                for k in range(cut):
                    if(newl[j][k] in gene):
                        newl[j][k]=-1
                        
                permu=[w for w in range(amount)]
                for k in range(len(newl[j])):
                    if newl[j][k] in permu:
                        permu.remove(newl[j][k])
                for k in range(len(newl[j])):
                    if newl[j][k]==-1:
                        newl[j][k]=permu.pop()
                
            st1=newl[0]
            st2=newl[1]
            st3=newl[2]
            st4=newl[3]
            
        return minpath      
                
            
            
xx= agent(5,0)
xx.TSP.show_graph_dist()
print()

print('hill-climbing:')
a= xx.hill_climbing(10)
print(a)
print(xx.calFinalCost_list(a))

print('\n genetic:')
b= xx.genetic(10)
print(b)
print(xx.calFinalCost_list(b))
```
